```python
import numpy as np
import pandas as pd
import sklearn
import seaborn
import matplotlib
```

```python
from sklearn.datasets import fetch_california_housing
var = fetch_california_housing()
var.keys()
```

```python
df = pd.DataFrame(var.data, columns=var.feature_names)
df
```

```python
df["MEDV"]=var.target
df.MEDV
```

```python
df.isnull().sum()
```

```python
matplotlib.pyplot.figure(figsize=(10, 6))
seaborn.boxplot(data=df)
```

```python
matplotlib.pyplot.figure(figsize=(10, 6))
seaborn.heatmap(data=df)
```

```python
y1=df.MEDV
x1=df.drop('MEDV',axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x1, y1, test_size=0.5,random_state = 5 )
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
pred= pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
import matplotlib.pyplot as plt
plt.plot(pred.Actual,pred.Actual)
plt.plot(pred.Actual[:100],pred.Predicted[:100],'o')
```

```python
from sklearn import metrics
metrics.mean_squared_error(pred.Actual,pred.Predicted)
```
